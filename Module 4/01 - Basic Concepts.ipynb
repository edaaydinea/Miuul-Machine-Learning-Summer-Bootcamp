## Variable Types

### Numerical Variables
  - ***Interval***
    - ,Non-zero, Equal Intervals, Order, Category
    - *Example*: Temperature in Fahrenheit - Celsius, SAT Scores, IQ, Year
  - ***Ratio***
    - True - Zero, Equal Intervals, Order, Category
    - *Example*: Age, Height, Weight, Temperature in Kelvin
  - ***Cardinal***
    - counted, discrete, quantity
    - *Example*: Coffee size (Small, Medium, Venti)


### Categorical Variables
  - ***Nominal***
    - Category
    - *Example*: Marital Status, Sex, Gender, Ethnicity, Blood type
  - ***Ordinal***
    - Order, Category
    - *Example*: Student Letter Grade, NFL Team Rankings

### Dependent Variables
- Target, Dependent, Output, Response

### Independent Variables
- Feature, Independent, Input, Column, Predictor, Explanatory



## Learning Types

Machine Learning

1. ***Supervised Learning (Denetimli Öğrenme)***
    -  If our labels are included in the data set, it is supervised learning. If there is a dependent variable, it is supervised learning.
    -  We look at the target in the dataset, while the features have these values, the target has this value. In this case, the algorithm tries to establish a relationship between these features and target. That's why, we called Supervised Learning. The algorithms learns by examining relationship.
2. ***Unsupervised Learning (Denetimsiz Öğrenme)***
    -   If we do not have labels in the data set, that is, we do not have a dependent (target) variable, it is unsupervised learning.
    -   If we do not have any relationship to examine between features and target, we called this learning type as unsupervised learning
    -   In this learning, we can get the relationship by clustering, segmenting etc. similar variables.
3. ***Reinforcement Learning (Pekiştirmeli Öğrenme)***
    - It is a way of learning by trial and error.

## Model Success Evaluation Methods

Our Goal: How successful are my predictions?

When the models we have built make predictions, we expect certain deviations in these predictions, (incorrect predictions are called residuals)

***Success Evaluation in Regression Models***

----

**Mean Squared Error**
- It is a measure of success evaluation. It provides the opportunity to evaluate the differences between the actual values ​​$y_{i}$ as a result of the model we have established and the estimated values ​​$\hat{y}_{i}$ obtained through the model.
- Always positive value.
- It can be said that estimators with an MSE value close to 0 perform better.
- MSE is the Mean error metric.

$$MSE = \frac{1}{n}\sum_{i=1}^{n}(y_{i}-\hat{y}_{i})^2$$

- Example:
    - $y_{i}$ : Actual Sales Values
    - $\hat{y}_{i}$ : Estimated Sales Values


----

**Root Mean Square Error**

$$\sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_{i}-\hat{y}_{i})^2}$$

- It is the standard deviation of the estimation errors (residues).
- That is, residuals are a measure of how far the regression line is from the data points.
- The RMSE value can vary from 0 to ∞.
- Negatively oriented scores, i.e. predictors with lower values, perform better.
- If the RMSE Value is zero, it indicates that the model has made no errors. RMSE has the advantage of penalizing major mistakes more, so it may be better suited to some situations.

----

**Mean Absolute Error**

$$\frac{1}{n}\sum_{i=1}^{n}\left | y_{i}-\hat{y}_{i}\right |$$

- It is the measure of the difference between two continuous variables.
- MAE is the average vertical distance between each actual value and the line that best fits the data.
- MAE is also the average horizontal distance between each data point and the best-fit line.
- MAE value can vary from 0 to ∞.
- Negatively oriented scores, i.e. predictors with lower values, perform better.

## Model Validation Methods

**Holdout Method (Sınama Seti Yönetimi)**

The original dataset is divided into Training Set and Test Set. The model is built on the Training Set, then it is tested by asking questions from the test set. This is how success is evaluated.

![Holdout Method](holdout_method1.png)

**K-Fold Cross Validation Method (K-Katlı Çapraz Doğrulama Yöntemi)**

Cross validation is a resampling procedure used to evaluate machine learning model on a limited data sample.

The general procedure is as follow:
1. Shuffle the dataset randomly
2. Split the dataset into k groups
3. For each unique group:
    1. Take the group as a hold out or test data set
    2. Take the remaining groups as a training data set
    3. Fit a model on the training set and evaluate it on the test set
    4. Retain the evaluation score and discard the model
4. Summarize the skill of the model using the sample of model evaluation scores

Each sample is given the opportunity to be used in the hold out se 1 time and used to train the model k-1 times.

We can also apply this method only on the training set.

## Bias - Variance Tradeoff

**Building a Model**
- It means to extract the relationship between the dependent and independent variable, the relationship. It means learning about the relationship.

**Overfitting**
- The model memorizes the data.
- High Variance
- Low training error | High test error

**Underfitting**
- It is the problem that the model cannot learn the data.
- High Bias
- High training error |High test error

**Optimal**
- It is the representation of the relationship that the data set carries.
- Low bias
- Low variance
- Low training error | Low test error

**Model Kurmak**
- Bağımlı ve bağımsız değişken arasındaki ilişkiyi , ilişkinin özütünü çıkarmak demektir. Bu ilişkiyi öğrenmek demektir.

**Overfitting**
- Modelin veriyi ezberlemesidir.
- Yüksek Varyans

**Underfitting**
- Modelin veriyi öğrenememe problemidir.

**Optimal**
- Veri setinin taşıdığı ilişkinin temsil edilmesidir.
- Düşük yanlılık
- Düşük varyans
- Düşük training hatası | Düşük test hatası

**How do yo know if you have fallen into overfitting?How is overfitting detected**
- The over-learning problem can be detected by evaluating the training set and test set together in the framework of Model Complexity and Prediction Error.
- Error changes in the training set and test set are examined, and it is said that excessive learning has started as of the point where these two errors begin to separate from each other.
- If we stop the model complexity, training time and iteration at the point where we detect excessive learning, we can avoid this problem.

**What is Model Complexity?**
- It is the refinement of the model. It is the strengthening of its features so that more detailed predictions can be made. It is the branching of trees in tree methods. 
- Complexity is not bad for a while, but it should not be complicated after a point, as it is memorized after a point.

**How to solve Overfitting?**
- There can be many answers to this. The size of the dataset can be increased. Feature selection can be made. Finally, the point where the errors of the training set and the test set differ from each other is selected as the optimum point. By stopping at this point, over-learning is prevented.

**Aşırı öğrenmeye düştüğünüzü nasıl anlarsınız? Aşırı Öğrenme nasıl tespit edilir?**
- Aşırı öğrenme sorunu, eğitim seti ve test setinin Model Karmaşıklığı ve Tahmin Hatası çerçevesinde birlikte değerlendirilmesi ile tespit edilebilir. 
- Eğitim seti ve test setindeki hata değişimleri incelenir, bu iki hatanın birbirinden ayrılmaya başladığı nokta itibari ile aşırı öğrenme başlamıştır denir.
- Aşırı öğrenmeyi tespit ettiğimiz noktada model karmaşıklığını, eğitim süresini iterasyonu durdurursak bu problemin önüne geçeriz.

**Model Karmaşıklığı nedir?**
- Modelin hassaslaştırılmasıdır. Daha detaylı tahmin yapılabilmesi için özelliklerinin kuvvetlendirilmesidir. Ağaç yöntemlerinde ağaçların dallandırılmasıdır. 
- Karmaşıklaştırma bir süreye kadar kötü değildir ama bir noktadan sonra ezberlenmeye gidildiği için bir noktadan sonra karmaşıklaştırılma yapılmamalıdır.


**Aşırı Öğrenme nasıl çözülür?**
- Bunun birçok yanıtı olabilir. Veri setinin boyutu arttırılabilir. Feature selection yapılabilir.En nihayetinde, eğitim seti ile test setinin hatalarının birbirlerinden ayrıldığı nokta seçilir optimum nokta olarak.  Bu noktada durulduğunda aşırı öğrenmenin önüne geçilmiş olunur.
